```python
import numpy as np
import matplotlib.pyplot as plt
from decimal import Decimal
```

## Gambler's Problem

### Using Value Iteration

### No unique optimal policy!!!

```python
N = 100 # max capital (0 is not used)

p = np.float64(0.4) # The probability of the coin coming up heads.

V = np.zeros([N+1]) #  containing two dummy states V[0]=0 and V[N]=1
a_opt = np.zeros(N)

V[N] = np.float64(1.0) # win if he has 100 stakes

isStable = 0

minError = np.float64(1e-25)

gamma = np.float64(1.0)

sweep = 0

while not isStable:
    
    

    sweep += 1
    
    isStable = 1
    
    for s in range(N):
        
        V_tmp = np.zeros([N])
        if not s==0:
            
            V_pre = V[s]
            
            for a in range(min(s,N-s)+1): # a is the stake, from 0 to min(s,100-s)
                
                sp_up = s + a
                sp_pu = s - a
           
                V_tmp[a] = p * ( gamma * V[sp_up]) + (1-p) * ( gamma * V[sp_pu])
                
            
#             V_tmp[0] = 0 # to avoid no action
            a_opt[s] = np.argmax(V_tmp)    
            V[s] = np.max(V_tmp)
            
            if abs(V_pre - V[s]) > minError:
                isStable = 0
                
                
        
        
    print(sweep)
    plt.plot(V)       

    
    
    
# Output optimal actions
for s in range(N):

    if not s==0:

        V_pre = V[s]
        V_tmp = np.zeros([N])
        
        for a in range(min(s,N-s)+1): # a is the stake, from 0 to min(s,100-s)
            sp_up = s + a
            sp_pu = s - a

            V_tmp[a] = p * ( gamma * V[sp_up]) + (1-p) * ( gamma * V[sp_pu])
        
#         print(V_tmp)
        V_tmp[0] = 0 # to avoid no action
        
        a_opt[s] = np.argmax(V_tmp)    

    

plt.show()
```

```python
plt.plot(V)
plt.show()
plt.plot(a_opt)
plt.show()
```
