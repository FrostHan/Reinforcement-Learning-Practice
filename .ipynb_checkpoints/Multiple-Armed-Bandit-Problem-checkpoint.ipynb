```python
import numpy as np
import matplotlib.pyplot as plt
```

## $\epsilon$-greedy action selection

```python
def reward_arm(n):
    if n==1:
        return np.random.normal(0.2,1)
    elif n==2:
        return np.random.normal(-0.7,1)
    elif n==3:
        return np.random.normal(1.5,1)
    elif n==4:
        return np.random.normal(0.5,1)
    elif n==5:
        return np.random.normal(1.3,1)
    elif n==6:
        return np.random.normal(-1.4,1)
    elif n==7:
        return np.random.normal(-0.2,1)
    elif n==8:
        return np.random.normal(-0.8,1)
    elif n==9:
        return np.random.normal(0.8,1)
    elif n==10:
        return np.random.normal(-0.6,1)
    
```

```python
epsilon = 0.1
r01 = np.zeros([1000,2000])
for episode in range(2000):
    

    Q = np.zeros([10])
    N = np.zeros([10])


    for step in range(1000):
        if np.random.rand()<epsilon:
            a = np.random.randint(1,11)
            r01[step, episode] = reward_arm(a)
            N[a-1] += 1 
            Q[a-1] = ((N[a-1]-1)*Q[a-1] + r01[step,episode])/N[a-1]
        else:
            a = np.argmax(Q) + 1
            r01[step, episode] = reward_arm(a)
            N[a-1] += 1 
            Q[a-1] = ((N[a-1]-1)*Q[a-1] + r01[step,episode])/N[a-1]
```
